```python
import scipy.integrate as sc
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion(t,y,omega_0,drag,c):
    x,v=y
    return np.array([v,omega_0**2*x-drag*v-(c*x**3)]) 
```

```python
sol=sc.solve_ivp(funcion,y0=np.array([1.,0.]),t_span=(0,100),args=[0.5,0.05,20],max_step=0.1)
```

```python
omega_0=0.5
energia=(sol.y[0]**2)*(omega_0**2)+(sol.y[1]**2)
```

```python
plt.plot(sol.t,sol.y[0],label="Posicion",alpha=0.5)
plt.plot(sol.t,sol.y[1],label="Velocidad")
plt.plot(sol.t,energia,label="Energa")
plt.grid()
plt.legend()
```
